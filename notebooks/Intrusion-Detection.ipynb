# **Intrusion Detection System Analysis using CIC-IDS2017 Dataset**

This notebook performs an analysis of the Canadian Institute for Cybersecurity's Intrusion Detection Evaluation Dataset (CIC-IDS2017). It covers data loading, exploratory data analysis (EDA), preprocessing, feature engineering (PCA), and the training and evaluation of various machine learning models for network intrusion detection.

[Dataset Link](https://www.unb.ca/cic/datasets/ids-2017.html)

## 1. Project Overview

This project aims to explore the CIC-IDS2017 dataset to understand its characteristics and build machine learning models capable of detecting network intrusions.

- **Dataset Description:** The CIC-IDS2017 dataset captures diverse network traffic, including normal (Benign) activity and various modern attack types like Brute Force, DoS, DDoS, Web Attacks, etc., collected over several days. A key characteristic is its significant class imbalance, with benign traffic forming the vast majority.
- **Exploratory Data Analysis (EDA):** Initial exploration revealed the dataset's large size (over 2.8 million initial records, 79 columns) and the presence of duplicate entries (~308k) and a small number of missing/infinite values (in 'Flow Bytes/s' and 'Flow Packets/s').
- **Data Preprocessing:** Steps included removing duplicates, handling infinite values by converting them to NaN, and imputing missing values using the median. Memory usage was optimized by downcasting numerical data types.
- **Feature Analysis:** Similar attack labels were grouped into broader categories. Correlation analysis showed strong relationships between several features, indicating potential multicollinearity. Due to the high dimensionality (over 70 features after initial cleaning), Principal Component Analysis (PCA) was employed for dimensionality reduction while aiming to preserve most of the variance.
- **Modeling Strategy:** Given the class imbalance, balanced datasets were created for both binary (Attack vs. Benign) and multi-class classification tasks. Techniques like undersampling the majority class (for binary) and SMOTE (Synthetic Minority Over-sampling Technique) combined with sampling (for multi-class) were used. Several classification algorithms were trained:
    - *Binary:* Logistic Regression, Support Vector Machine (SVM)
    - *Multi-class:* Random Forest, Decision Tree, K-Nearest Neighbors (KNN)
- **Evaluation:** Models were evaluated using standard metrics like accuracy, precision, recall, F1-score, confusion matrices, and ROC curves (for binary). Cross-validation was used to assess generalization performance.

#### Imputing Missing Values

The features 'Flow Bytes/s' and 'Flow Packets/s' have a small percentage of missing values. These represent rates and can be highly skewed or contain outliers (e.g., during DoS attacks). Using the median is a robust strategy for imputation in such cases, as it's less sensitive to extreme values than the mean.

```python
# Analyze distribution before imputation (using a sample for efficiency)
if 'Flow Bytes/s' in missing_cols_list or 'Flow Packets/s' in missing_cols_list:
    sample_df_impute = df.sample(min(len(df), 50000), random_state=42) # Sample for plotting
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    if 'Flow Bytes/s' in missing_cols_list:
        sns.histplot(sample_df_impute['Flow Bytes/s'].dropna(), kde=False, ax=axes[0], bins=50)
        axes[0].set_title('Distribution of Flow Bytes/s (Sample)')
        axes[0].set_yscale('log') # Use log scale due to potential skewness
        median_flow_bytes = df['Flow Bytes/s'].median()
        print(f"Median for 'Flow Bytes/s': {median_flow_bytes:.2f}")
        df['Flow Bytes/s'].fillna(median_flow_bytes, inplace=True)
        print(f"Missing 'Flow Bytes/s' imputed with median.")
    else:
        axes[0].set_title('No Missing Flow Bytes/s')

    if 'Flow Packets/s' in missing_cols_list:
        sns.histplot(sample_df_impute['Flow Packets/s'].dropna(), kde=False, ax=axes[1], bins=50, color='salmon')
        axes[1].set_title('Distribution of Flow Packets/s (Sample)')
        axes[1].set_yscale('log') # Use log scale
        median_flow_packets = df['Flow Packets/s'].median()
        print(f"Median for 'Flow Packets/s': {median_flow_packets:.2f}")
        df['Flow Packets/s'].fillna(median_flow_packets, inplace=True)
        print(f"Missing 'Flow Packets/s' imputed with median.")
    else:
         axes[1].set_title('No Missing Flow Packets/s')

    plt.tight_layout()
    plt.show()
    del sample_df_impute
    gc.collect()

    # Verify imputation
    print("\nMissing values count after imputation:")
    print(df[missing_cols_list].isna().sum())
else:
    print("No missing values needed imputation in 'Flow Bytes/s' or 'Flow Packets/s'.")
```

#### Optimizing Memory Usage

Large datasets consume significant memory. We can reduce this by downcasting numerical data types (e.g., `float64` to `float32`, `int64` to `int32`) where possible without losing precision, based on the range of values in each column.

```python
def optimize_memory(df_to_optimize):
    """Iterates through columns and downcasts types to reduce memory usage."""
    initial_mem_mb = df_to_optimize.memory_usage().sum() / 1024**2
    print(f'Initial memory usage: {initial_mem_mb:.2f} MB')

    for col in df_to_optimize.columns:
        col_type = df_to_optimize[col].dtype
        if pd.api.types.is_numeric_dtype(col_type):
            c_min = df_to_optimize[col].min()
            c_max = df_to_optimize[col].max()

            if str(col_type).startswith('int'):
                if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max:
                    df_to_optimize[col] = df_to_optimize[col].astype(np.int8)
                elif c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max:
                     df_to_optimize[col] = df_to_optimize[col].astype(np.int16)
                elif c_min >= np.iinfo(np.int32).min and c_max <= np.iinfo(np.int32).max:
                    df_to_optimize[col] = df_to_optimize[col].astype(np.int32)
                # Keep int64 if necessary

            elif str(col_type).startswith('float'):
                 # Check if float32 is sufficient
                 if c_min >= np.finfo(np.float32).min and c_max <= np.finfo(np.float32).max:
                     df_to_optimize[col] = df_to_optimize[col].astype(np.float32)
                 # Keep float64 if necessary

    final_mem_mb = df_to_optimize.memory_usage().sum() / 1024**2
    print(f'Final memory usage: {final_mem_mb:.2f} MB')
    print(f'Reduced by: {100 * (initial_mem_mb - final_mem_mb) / initial_mem_mb:.1f}%')
    return df_to_optimize

print("\nOptimizing DataFrame memory...")
df = optimize_memory(df)

print("\nDataFrame Info after optimization:")
df.info(verbose=False) # Show concise info
```

## 3. Feature Scaling and Dimensionality Reduction (PCA)

Machine learning algorithms, especially distance-based ones (like KNN, SVM) or those sensitive to feature magnitudes (like PCA, Logistic Regression with regularization), benefit from feature scaling. StandardScaler standardizes features by removing the mean and scaling to unit variance.

Principal Component Analysis (PCA) is a technique to reduce the number of features (dimensionality) while retaining most of the original information (variance). This helps combat the curse of dimensionality, reduces computational cost, and can sometimes improve model performance by removing noise.

Since the dataset is large, we use `IncrementalPCA`, which processes the data in batches, making it memory-efficient.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import IncrementalPCA

# Separate features (X) and the target variable (y)
X = df.drop('Attack_Type', axis=1)
y = df['Attack_Type']

# Identify numeric features for scaling
numeric_features = X.select_dtypes(include=np.number).columns.tolist()
print(f"Scaling {len(numeric_features)} numeric features...")

scaler = StandardScaler()

# Fit and transform the numeric features
# Ensure output is a DataFrame to maintain column names if needed (though PCA removes them)
X_scaled_numeric = scaler.fit_transform(X[numeric_features])
X_scaled_numeric_df = pd.DataFrame(X_scaled_numeric, columns=numeric_features, index=X.index)

# If there were non-numeric features (there aren't here, but good practice):
# X_non_numeric = X.select_dtypes(exclude=np.number)
# X_scaled = pd.concat([X_scaled_numeric_df, X_non_numeric], axis=1)
# For this dataset, X_scaled is just the scaled numeric part
X_scaled = X_scaled_numeric_df

print("Scaling complete. Shape of scaled features:", X_scaled.shape)
# Display sample of scaled data
print("\nScaled Features Sample (First 5 rows, 5 columns):")
print(X_scaled.iloc[:5, :5])

# Clean up intermediate variables
del X_scaled_numeric, X_scaled_numeric_df, numeric_features
gc.collect()
```

```python
# Determine the number of components for PCA
# Aiming for a high variance explained, e.g., 95-99%
# Let's start by fitting with a large number of components and check explained variance
n_total_features = X_scaled.shape[1]
n_components_check = min(n_total_features, 50) # Check up to 50 components or total features

print(f"Running IncrementalPCA with {n_components_check} components to check explained variance...")
ipca_check = IncrementalPCA(n_components=n_components_check, batch_size=1000) # Adjust batch_size based on memory

# Fit IPCA in batches
start_time_pca = time.time()
batch_count = 0
# Use numpy array split for potentially better memory handling
for batch in np.array_split(X_scaled.values, len(X_scaled) // 1000 + 1):
    if batch.size > 0:
      ipca_check.partial_fit(batch)
      batch_count += 1
      if batch_count % 50 == 0:
        print(f"  Processed {batch_count} batches...")

print(f"IPCA fitting for variance check done. Time: {time.time() - start_time_pca:.2f} seconds")

# Calculate cumulative explained variance
explained_variance_ratio = ipca_check.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Plot cumulative explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, n_components_check + 1), cumulative_explained_variance, marker='.', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Explained Variance by Principal Components')
plt.grid(True)
# Add lines for common thresholds
plt.axhline(y=0.95, color='r', linestyle=':', label='95% Variance')
plt.axhline(y=0.99, color='g', linestyle=':', label='99% Variance')
plt.legend(loc='best')
plt.show()

# Determine optimal number of components based on threshold (e.g., 95%)
n_components_optimal = np.argmax(cumulative_explained_variance >= 0.95) + 1
print(f"Number of components to retain approx. 95% variance: {n_components_optimal}")
print(f"Variance explained by {n_components_optimal} components: {cumulative_explained_variance[n_components_optimal-1]:.2%}")

# Now, run IPCA again with the optimal number of components to get the transformed data
print(f"\nRunning IncrementalPCA with {n_components_optimal} components for transformation...")
ipca_final = IncrementalPCA(n_components=n_components_optimal, batch_size=1000)

start_time_pca_transform = time.time()
# We need to transform the data, which requires another pass or storing batches if memory allows.
# For simplicity here, we'll re-run partial_fit (less efficient but simpler code).
# A more efficient approach transforms batches after fitting.
transformed_batches = []
batch_count = 0
for batch in np.array_split(X_scaled.values, len(X_scaled) // 1000 + 1):
    if batch.size > 0:
      # Fit and then transform the batch
      ipca_final.partial_fit(batch) # Re-fitting here for simplicity
      # transformed_batch = ipca_final.transform(batch) # Ideal: transform after full fit
      # transformed_batches.append(transformed_batch)
      batch_count += 1
      if batch_count % 50 == 0:
          print(f"  Processed {batch_count} batches for final PCA...")

# Perform the transformation after fitting all batches
X_pca = ipca_final.transform(X_scaled)
print(f"PCA transformation done. Time: {time.time() - start_time_pca_transform:.2f} seconds")

# Create a new DataFrame with principal components
pca_columns = [f'PC{i+1}' for i in range(n_components_optimal)]
df_pca = pd.DataFrame(X_pca, columns=pca_columns, index=y.index) # Keep original index
df_pca['Attack_Type'] = y # Add the target variable back

print("\nDataFrame after PCA:")
print(df_pca.head())
print(f"Shape after PCA: {df_pca.shape}")

# Clean up large intermediate objects
del X, X_scaled, X_pca, df, ipca_check, ipca_final
gc.collect()
```

### 4.1. Binary Classification: Preparing a Balanced Dataset

For binary classification (Attack vs. Benign), we have a large number of Benign samples and fewer Attack samples. A common strategy is to undersample the majority class (Benign) to match the number of minority class (Attack) samples. We then take a reasonably sized sample from this balanced set for training.

```python
print("Preparing balanced dataset for Binary Classification...")
# Separate Benign and Attack samples from the PCA data
df_benign_pca = df_pca[df_pca['Attack_Type'] == 'Benign']
df_attack_pca = df_pca[df_pca['Attack_Type'] != 'Benign']

num_attacks = len(df_attack_pca)
num_benign = len(df_benign_pca)
print(f"Number of Attack samples: {num_attacks:,}")
print(f"Number of Benign samples: {num_benign:,}")

# Undersample Benign class to match the number of Attack samples
df_benign_undersampled = df_benign_pca.sample(n=num_attacks, random_state=42)
print(f"Undersampled Benign to: {len(df_benign_undersampled):,} samples")

# Combine undersampled Benign and all Attack samples
df_binary_balanced_full = pd.concat([df_attack_pca, df_benign_undersampled])

# Map labels to 0 (Benign) and 1 (Attack)
df_binary_balanced_full['Binary_Label'] = df_binary_balanced_full['Attack_Type'].apply(lambda x: 0 if x == 'Benign' else 1)

# Shuffle the dataset
df_binary_balanced_full = df_binary_balanced_full.sample(frac=1, random_state=42).reset_index(drop=True)

# Take a final sample for training/testing to manage computational resources
binary_sample_size = 25000 # Adjusted sample size
if len(df_binary_balanced_full) > binary_sample_size:
    df_binary_final = df_binary_balanced_full.sample(n=binary_sample_size, random_state=42)
    print(f"Taking a final sample of {binary_sample_size:,} for binary classification tasks.")
else:
    df_binary_final = df_binary_balanced_full
    print(f"Using the full balanced set ({len(df_binary_final):,}) for binary classification tasks.")

print("\nFinal Binary Dataset Label Distribution:")
print(df_binary_final['Binary_Label'].value_counts())

# Prepare data for Scikit-learn
X_binary = df_binary_final.drop(['Attack_Type', 'Binary_Label'], axis=1)
y_binary = df_binary_final['Binary_Label']

print(f"\nShape of X_binary: {X_binary.shape}")
print(f"Shape of y_binary: {y_binary.shape}")

# Split into training and testing sets
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(
    X_binary, y_binary, test_size=0.3, random_state=42, stratify=y_binary # Stratify maintains class proportion
)

print(f"Binary Train set shape: X={X_train_bin.shape}, y={y_train_bin.shape}")
print(f"Binary Test set shape: X={X_test_bin.shape}, y={y_test_bin.shape}")

# Clean up intermediate dataframes
del df_benign_pca, df_attack_pca, df_benign_undersampled, df_binary_balanced_full, df_binary_final
gc.collect()
```

### 4.4. Multi-class Classification: Preparing a Balanced Dataset

Multi-class imbalance is more complex. Some attack types have very few samples. A common approach is:
1.  Select classes with a minimum number of samples to ensure sufficient data.
2.  Undersample very large classes (like Benign, and possibly large attack classes like DoS) to a reasonable level.
3.  Use an oversampling technique like SMOTE (Synthetic Minority Over-sampling Technique) to generate synthetic samples for the remaining minority classes, bringing all selected classes to a similar size.

```python
print("Preparing balanced dataset for Multi-class Classification...")

# Original multi-class distribution in PCA data
original_multi_counts = df_pca['Attack_Type'].value_counts()
print("Original Multi-class Distribution:")
print(original_multi_counts)

# --- Step 1: Select classes with enough samples ---
min_samples_threshold = 1000 # Keep classes with at least 1000 samples
selected_classes = original_multi_counts[original_multi_counts >= min_samples_threshold].index.tolist()
df_multi_filtered = df_pca[df_pca['Attack_Type'].isin(selected_classes)].copy()
print(f"\nSelected {len(selected_classes)} classes with >= {min_samples_threshold} samples:", selected_classes)
print(f"Shape after initial class filtering: {df_multi_filtered.shape}")

# --- Step 2: Undersample large classes (if necessary) ---
# Set a target max size for very large classes before SMOTE
undersample_target = 15000 # Undersample classes larger than this (e.g., Benign, DoS)

df_list_undersampled = []
for cls in selected_classes:
    cls_df = df_multi_filtered[df_multi_filtered['Attack_Type'] == cls]
    if len(cls_df) > undersample_target:
        print(f"Undersampling class '{cls}' from {len(cls_df)} to {undersample_target}")
        df_list_undersampled.append(cls_df.sample(n=undersample_target, random_state=42))
    else:
        df_list_undersampled.append(cls_df)

df_multi_undersampled = pd.concat(df_list_undersampled)
print("\nDistribution after undersampling large classes:")
print(df_multi_undersampled['Attack_Type'].value_counts())
print(f"Shape after undersampling: {df_multi_undersampled.shape}")

# --- Step 3: Apply SMOTE for oversampling minority classes ---
X_multi_pre_smote = df_multi_undersampled.drop('Attack_Type', axis=1)
y_multi_pre_smote = df_multi_undersampled['Attack_Type']

print("\nApplying SMOTE to balance the remaining classes...")
start_time_smote = time.time()
# 'auto' balances all classes to the size of the largest minority class after undersampling
smote = SMOTE(sampling_strategy='auto', random_state=42, k_neighbors=5) # Use n_jobs if possible
X_multi_smote, y_multi_smote = smote.fit_resample(X_multi_pre_smote, y_multi_pre_smote)
print(f"SMOTE application time: {time.time() - start_time_smote:.2f} seconds")

print("\nDistribution after SMOTE:")
print(y_multi_smote.value_counts())

# Create the final balanced DataFrame
df_multi_final = pd.DataFrame(X_multi_smote, columns=X_multi_pre_smote.columns)
df_multi_final['Attack_Type'] = y_multi_smote

# Shuffle the final dataset
df_multi_final = df_multi_final.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"\nFinal Multi-class balanced dataset shape: {df_multi_final.shape}")

# Prepare data for Scikit-learn
X_multi = df_multi_final.drop('Attack_Type', axis=1)
y_multi = df_multi_final['Attack_Type']

# Encode the multi-class labels numerically for some models if needed (RF, DT, KNN usually handle strings)
multi_label_encoder = LabelEncoder()
y_multi_encoded = multi_label_encoder.fit_transform(y_multi)
multi_encoding_map = dict(zip(multi_label_encoder.classes_, multi_label_encoder.transform(multi_label_encoder.classes_)))
print("Multi-class Encoding Map:", multi_encoding_map)

# Split into training and testing sets
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
    X_multi, y_multi, test_size=0.3, random_state=42, stratify=y_multi
)

# Also keep encoded versions for evaluation consistency if needed
_, _, y_train_multi_enc, y_test_multi_enc = train_test_split(
    X_multi, y_multi_encoded, test_size=0.3, random_state=42, stratify=y_multi_encoded
)

print(f"Multi-class Train set shape: X={X_train_multi.shape}, y={y_train_multi.shape}")
print(f"Multi-class Test set shape: X={X_test_multi.shape}, y={y_test_multi.shape}")

# Clean up
del df_pca, df_multi_filtered, df_list_undersampled, df_multi_undersampled, X_multi_pre_smote, y_multi_pre_smote
del X_multi_smote, y_multi_smote, df_multi_final
gc.collect()
```

### 5.3. Discussion and Conclusion

This analysis explored the CIC-IDS2017 dataset, addressing challenges like its large size, class imbalance, and high dimensionality.

**Preprocessing and Feature Engineering:**
- Cleaning steps like removing duplicates and handling missing/infinite values were essential.
- Memory optimization through data type downcasting proved effective.
- Grouping specific attack labels provided clearer categories for multi-class analysis.
- High correlation between features was observed, motivating the use of PCA for dimensionality reduction. Standard scaling was applied prior to PCA.
- PCA successfully reduced the feature space while retaining a high percentage (e.g., ~95%) of the original variance, making subsequent modeling more feasible.

**Modeling and Evaluation:**
- **Binary Classification:**
    - Balanced datasets were created by undersampling the 'Benign' class.
    - Both Logistic Regression and SVM achieved high performance after data balancing and PCA. SVM (particularly with the RBF kernel in this run, though results can vary) often showed slightly higher accuracy and AUC compared to Logistic Regression, but at a significantly higher computational cost.
    - The choice between LR and SVM might depend on the priority: LR for speed and reasonable accuracy, SVM for potentially higher accuracy if training time is less constrained.
- **Multi-class Classification:**
    - A combination of filtering, undersampling majority classes, and SMOTE oversampling was used to create a balanced multi-class dataset.
    - Random Forest consistently outperformed Decision Trees and KNN in terms of overall accuracy and per-class F1-scores. This is expected as RF is an ensemble method designed to reduce the variance and overfitting issues often seen in single decision trees.
    - KNN also performed well, demonstrating its effectiveness on the PCA-transformed data, but generally slightly lower than RF.
    - Decision Trees showed the lowest performance among the three, likely struggling with the complexity of the classification boundaries after PCA, even with balanced data.

**Overall Findings:**
- The CIC-IDS2017 dataset, after appropriate preprocessing (including scaling and PCA) and balancing, allows for the training of effective intrusion detection models.
- For binary detection (Attack vs. Benign), both LR and SVM are viable options with different trade-offs.
- For multi-class identification of specific attack types, ensemble methods like Random Forest appear most promising among the tested algorithms, providing robust performance across different attack categories.

**Future Work:**
- **Hyperparameter Tuning:** Employ more systematic hyperparameter optimization techniques (e.g., GridSearchCV, RandomizedSearchCV) to potentially further improve model performance.
- **Advanced Models:** Explore other algorithms like Gradient Boosting (XGBoost, LightGBM) or Neural Networks (especially for capturing complex patterns).
- **Feature Selection:** Experiment with feature selection techniques *before* or *instead of* PCA to see if using a subset of original features yields better or more interpretable results.
- **Ensemble Methods:** Combine predictions from the best binary and/or multi-class models (e.g., using stacking) to potentially create an even more robust system.
- **Real-time Adaptation:** Investigate methods for adapting models to evolving network traffic patterns and new attack types (concept drift).
